```python
# Better process the dataset
import re

def remove_hyperlinks(text):
    # Regular expression pattern to match URLs
    url_pattern = r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+'

    try:
      # Replace URLs with an empty string
      cleaned_text = re.sub(url_pattern, '', text)
    except Esxception as e:
      print("Parsing exception raised:\t{}".format(e))
      return text

    return cleaned_text

input_file_path = '/content/drive/MyDrive/processed_mastodon_dataset_big.txt'    # Path to the input text file
output_file_path = '/content/drive/MyDrive/parsed_processed_mastodon_dataset_big.txt'  # Path to the output file

with open(input_file_path, 'r', encoding='utf-8') as input_file:
    with open(output_file_path, 'w', encoding='utf-8') as output_file:
        for line in input_file:
            # Remove hyperlinks from the current line
            processed_line = remove_hyperlinks(line)

            # Write the processed line to the output file
            output_file.write(processed_line)
```

```python
# Since the BERTopic object accepts a list of strings as an input, the following snippet transforms the dataset (which is in .txt format) into said string list
docs = []

with open('/content/drive/MyDrive/parsed_processed_mastodon_dataset_big.txt', 'r') as file:
    for line in file:
        docs.append(line.strip())

# ~ head 5
for i in range(5):
  print(docs[i])

# Show the list length, just to check
print("String list length:\t{}".format(len(docs)))
```

```python
# This is the most computationally intensive portion of the notebook, the trainig phase
# Without the hardware acceleration so kindly offered by google, it would've definitely
# Taken MUCH longer to train this model
from bertopic import BERTopic

topic_model = BERTopic(embedding_model="all-MiniLM-L6-v2", verbose=True)

# topic_model.get_topic_info()
topics, probs = topic_model.fit_transform(docs)
topic_model.get_topic_info()
```

```python
# After the training is complete, an important step is to export the trined model
# So that I may use it in the future

embedding_model = "sentence-transformers/all-MiniLM-L6-v2"

topic_model.save("/content/drive/MyDrive/", serialization="pytorch", save_ctfidf=True, save_embedding_model=embedding_model)
```

```python
# Alternative save format
import joblib

model_filename = "/content/drive/MyDrive/models/bertopic_mastodon_model.pkl"
joblib.dump(topic_model, model_filename)
```
